# Unified API Demo - ShortcutDetector

This notebook demonstrates the `ShortcutDetector` API for detecting shortcuts in embeddings.

**What you'll learn:**
- How to detect shortcuts using HBAC, Probe, and Statistical methods
- How to generate a comprehensive report with visualizations
- How to export results to CSV and PDF

```python
import numpy as np

from shortcut_detect import ShortcutDetector
from shortcut_detect import generate_linear_shortcut, generate_no_shortcut

print("✅ Imports successful!")
```

```python
# Create output directories for saving reports and visualizations
import os
os.makedirs('../output', exist_ok=True)
os.makedirs('../output/csv_exports', exist_ok=True)
print("✅ Output directories ready")
```

## Generate Synthetic Data

Create embeddings with known shortcuts in the first 5 dimensions.

```python
# Generate data with shortcuts
embeddings, labels = generate_linear_shortcut(
    n_samples=1000,
    embedding_dim=128,
    shortcut_dims=5,  # First 5 dimensions have shortcuts
    seed=42
)

print(f"Generated embeddings: {embeddings.shape}")
print(f"Class distribution: {np.bincount(labels)}")
```

## Run Shortcut Detection

Detect shortcuts using all three methods: HBAC, Probe, and Statistical tests.

```python
# Create detector with all methods
detector = ShortcutDetector(
    methods=['hbac', 'probe', 'statistical'],
    seed=42
)

# Fit on data
detector.fit(embeddings, labels)
```

### View Summary

```python
# Print text summary
print(detector.summary())
```

## Generate Comprehensive Report

Generate a single comprehensive report with:
- All detection results (HBAC, Probe, Statistical)
- Visualizations (PCA, t-SNE, dimension importance, heatmaps)
- CSV exports for further analysis
- PDF report for sharing

```python
# Generate ONE comprehensive report with everything
detector.generate_report(
    output_path='../output/comprehensive_shortcut_report.html',
    format='html',
    include_visualizations=True,  # Include all plots
    export_csv=True,              # Export CSV files
    csv_dir='../output/csv_exports'  # CSV directory
)

print("\n✅ Comprehensive report generated!")
print("   - HTML report: ../output/comprehensive_shortcut_report.html")
print("   - CSV exports: ../output/csv_exports/")
```

## Optional: Generate PDF Report

You can also generate a PDF version of the same report.

```python
# Optional: Generate PDF report (requires weasyprint + system dependencies)
# On macOS: brew install pango gdk-pixbuf libffi
# On Ubuntu: apt-get install libpango-1.0-0 libpangocairo-1.0-0

try:
    detector.generate_report(
        output_path='../output/comprehensive_shortcut_report.pdf',
        format='pdf',
        include_visualizations=True
    )
    print("✅ PDF report generated: ../output/comprehensive_shortcut_report.pdf")
except OSError as e:
    if "libgobject" in str(e) or "library" in str(e).lower():
        print("⚠️  PDF generation skipped - missing system dependencies")
        print("   To enable PDF export, install required libraries:")
        print("   • macOS: brew install pango gdk-pixbuf libffi")
        print("   • Ubuntu: apt-get install libpango-1.0-0 libpangocairo-1.0-0")
        print("\n   HTML reports work without these dependencies!")
    else:
        raise
```
